```python
# Impor library yang diperlukan
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Memuat dataset Iris
iris = load_iris()
X = iris.data  # Fitur-fitur
```

```python
# Standarisasi fitur-fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Inisialisasi model KMeans untuk clustering dengan 3 kelompok (clusters)
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
# Melatih model menggunakan data yang telah distandarisasi
kmeans.fit(X_scaled)
```

```python
# Mendapatkan label klaster untuk setiap data
labels = kmeans.labels_
```

```python
# Mendapatkan koordinat pusat klaster
centroids = kmeans.cluster_centers_
```

```python
# Visualisasi hasil klasterisasi
plt.figure(figsize=(8, 6))
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, s=50, cmap='viridis', alpha=0.7, label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, c='r', label='Centroids')
```

```python
plt.title('KMeans Clustering')
plt.xlabel('Sepal Length (scaled)')
plt.ylabel('Sepal Width (scaled)')
plt.legend()
plt.show()
```
